### Sistemazione dei valori della colonna "Data anno corrente"

```python
#Costruzione di una funzione che estrae da ogni data il giorno in stringa. 
#Questo servirà per proseguire il pre-processing dei dati mancanti utilizzando KNIME.
def giorno_settimana(colonna):
    
    if "Do" in colonna:
        valore = "Domenica"
        colonna = valore
        
    if "Lu" in colonna:
        valore = "Lunedi"
        colonna = valore
    if "Ma" in colonna:
        valore = "Martedi"
        colonna = valore
        
    if "Me" in colonna:
        valore = "Mercoledi"
        colonna = valore
        
    if "Gi" in colonna:
        valore = "Giovedi"
        colonna = valore
        
    if "Ve" in colonna:
        valore = "Venerdi"
        colonna = valore
        
    if "Sa" in colonna:
        valore = "Sabato"
        colonna = valore
        
    
    return colonna
    
risto["Giorno settimana"] = risto['Data anno corrente'].apply(giorno_settimana)
risto
```

```python
#sistemazione data anno corrente parte 1 
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Do ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Lu ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Ma ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Me ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Gi ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Ve ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Sa ","")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("Do ","")
```

```python
#sistemazione data anno corrente parte 2
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("gen ","01 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("feb ","02 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("mar ","03 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("apr ","04 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("mag ","05 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("giu ","06 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("lug ","07 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("ago ","08 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("set ","09 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("ott ","10 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("nov ","11 ")
risto["Data anno corrente"] = risto["Data anno corrente"].str.replace("dic ","12 ")

```
